```python
import sys
from pathlib import Path
import subprocess
import numpy as np
import matplotlib.pyplot as plt

PROJECT_ROOT = Path.cwd().resolve().parents[0]
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

FIG_DIR = PROJECT_ROOT / 'outputs' / 'figures'
DATA_DIR = PROJECT_ROOT / 'outputs' / 'data'
```

```python
subprocess.run([sys.executable, str(PROJECT_ROOT / 'scripts' / 'run_monopole.py')], check=True)
subprocess.run([sys.executable, str(PROJECT_ROOT / 'scripts' / 'run_parameter_sweep.py')], check=True)
subprocess.run([sys.executable, str(PROJECT_ROOT / 'scripts' / 'run_dipole.py')], check=True)
subprocess.run([sys.executable, str(PROJECT_ROOT / 'scripts' / 'run_detailed_source_placeholder.py')], check=True)
```

```python
baseline = np.load(DATA_DIR / 'monopole_baseline.npz')
sweep = np.load(DATA_DIR / 'parameter_sweep.npz', allow_pickle=True)
dipole = np.load(DATA_DIR / 'dipole_baseline.npz')
baseline['phi_elec'], dipole['phi_elec']
```

```python
gallery = [
    'geometry_schematic.png',
    'axisymmetric_field_map.png',
    'sweep_heatmap.png',
    'mechanism_curves.png',
    'dipole_geometry_3d.png',
    'dipole_slice_view.png',
    'dipole_clipped_view.png',
]

fig, axes = plt.subplots(len(gallery), 1, figsize=(9, 4 * len(gallery)))
for ax, name in zip(axes, gallery):
    ax.imshow(plt.imread(FIG_DIR / name))
    ax.set_title(name)
    ax.axis('off')
plt.tight_layout()
plt.show()
```
